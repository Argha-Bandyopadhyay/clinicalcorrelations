```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

## high-sensitivity

```python
disease_mean = 500
disease_sd = 100
normal_sd = 100

z_disease = st.norm.ppf(0.05)
threshold = (z_disease * disease_sd) + disease_mean

z_normal = st.norm.ppf(0.8)
normal_mean = threshold - (z_normal * normal_sd)

min = 0
max = 800
values = np.arange(min, max)
disease_pdf = st.norm.pdf(values, disease_mean, disease_sd)
normal_pdf = st.norm.pdf(values, normal_mean, normal_sd)
plt.plot(values, disease_pdf)
plt.plot(values, normal_pdf)
plt.xticks([])
plt.yticks([])
plt.ylabel('probability density')
plt.xlabel('hs-Troponin')
plt.axvline(threshold, c = 'r')

mask_values = np.arange(min, threshold)
mask_pdf = st.norm.pdf(mask_values, disease_mean, disease_sd)
plt.fill_between(mask_values, mask_pdf, color = '#1f77b4', alpha = 0.25)
plt.text(x = values[np.argmax(disease_pdf)], y = np.max(disease_pdf)/2, s = "myocardial\n infarction", color = '#1f77b4', ha = 'center', va = 'center')

mask_values = np.arange(threshold, max)
mask_pdf = st.norm.pdf(mask_values, normal_mean, normal_sd)
plt.fill_between(mask_values, mask_pdf, color = '#ff7f0e', alpha = 0.25)
plt.text(x = values[np.argmax(normal_pdf)], y = np.max(normal_pdf)/2, s = "healthy", color = '#ff7f0e', ha = 'center', va = 'center')
plt.savefig('C:/Users/argha/Documents/NYU/2 - Clerkships/sens_spec.png')
plt.show()
```

```python
ratio = disease_pdf / normal_pdf
disease_likely = ratio > 1
critical_mask = np.append(0, np.diff(disease_likely))
critical = values[critical_mask == 1]
below_threshold = int(threshold - (max - min)/10)
above_threshold = int(threshold + (max - min)/10)
window = np.arange(below_threshold, above_threshold)
plt.plot(values[window], ratio[window], c = '#2ca02c')
plt.xticks([])
plt.yticks([])
plt.ylabel('likelihood ratio')
plt.xlabel('hs-Troponin')
plt.axvline(threshold, c = 'r')
plt.axvline(critical, c = 'r', ls = '--')
plt.axhline(1, c = 'k')
plt.text(x = threshold + 20, y = 1.5, s = "positive\n for MI\n but\n healthy\n more\n likely", color = 'r', ha = 'center', va = 'center')
plt.savefig('C:/Users/argha/Documents/NYU/2 - Clerkships/sens_dynamic_lr.png')
plt.show()
```
